Data Science Internship – Task 1

## 1. Load Dataset

```
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('student-mat.csv', sep=';')
df.head()
```

## 2. Explore & Clean Data

```
print(df.isnull().sum())
print('Duplicates:', df.duplicated().sum())
df=df.drop_duplicates()
print(df.shape)
print(df.dtypes)
```

## 3. Analysis Questions

```
print('Average G3:', 10.415189873417722)
print('Students scoring >15:', 40)
print('Correlation studytime vs G3:', 0.09781968965319626)
print({'F': 9.966346153846153, 'M': 10.914438502673796})
```

## 4. Visualizations

```
df['G3'].hist()
plt.title('Histogram of Final Grades')
plt.show()

plt.scatter(df['studytime'], df['G3'])
plt.xlabel('Study Time')
plt.ylabel('Final Grade (G3)')
plt.show()

avg=df.groupby('sex')['G3'].mean()
avg.plot(kind='bar')
plt.title('Average Score by Gender')
plt.show()
```
